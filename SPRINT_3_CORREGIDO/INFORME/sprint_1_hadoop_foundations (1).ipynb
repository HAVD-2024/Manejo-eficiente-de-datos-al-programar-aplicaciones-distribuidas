## 2. Hadoop

### 2.1 ¿Qué es Apache Hadoop?

![title](./images/hadoop_logo.png)

**Figura 1: Logo de Apache Hadoop**

El software [Apache Hadoop](https://hadoop.apache.org) es un framework de código abierto que 
se utiliza para almacenar y procesar eficientemente conjuntos de datos muy grandes.

Para ello, Hadoop aprovecha las ventajas que ofrece el **cómputo distribuido** que no es más que un modelo de procesamiento de datos en el que las tareas y cargas de trabajo se dividen y distribuyen entre múltiples computadoras conectadas en una red. Dichan computadoras, también llamadas *nodos*, trabajan de manera coordinada para resolver problemas y completar tareas de manera más eficiente y rápida que si se utilizara una sola máquina.

El enfoque anterior posibilita que Hadoop pueda manejar tareas que involucran cantidades masivas de datos con eficacia y eficiencia y lo hace capaz de escalar verticalmente desde una computadora hasta miles de computadoras agrupadas en clústeres.

Cabe destacar que fue desarrollado inicialmente por Doug Cutting y Mike Cafarella en 2006, y patrocinado por posteriormente por la Apache Software Foundation. Además, la implementación está basada en el lenguaje de programación `Java`.

Antes de continuar, recomendamos ampliamente las siguientes referecias profundizar a detalle sobre Hadoop:

### 2.2 ¿Cómo funciona de Hadoop?

Cómo se ha dicho Hadoop opera sobre una arquitectura distribuida que permite dividir grandes conjuntos de datos en bloques más pequeños, los cuales se distribuyen y procesan en paralelo a través de múltiples nodos en un clúster. Esto se logra mediante varios componentes principales:

* **Hadoop Distributed File System (HDFS):** es el componente principal del ecosistema de Hadoop. Se trata un sistema de archivos distribuido en múltiples computadoras que se ejecuta en hardware estándar. Lo anterior facilita gestionar el almacenamiento de datos en grandes volúmenes ya que por su diseño los datos son accedidos de manera eficiente a través de un clúster de nodos (es decir, con alta escalabilidad y remdimiento, tolerancia a fallos así como flexibilidad).

* **Yet another resource negotiator (YARN):** es una plataforma de administración que se encarga de administrar los recursos de procesamiento en clústeres y usarlos para programar las aplicaciones de los usuarios; es decir cumple la función de programar y asigna recursos en el sistema Hadoop.

* **MapReduce:** Es un modelo de programación para el procesamiento de datos mediante algoritmos de procesamiento distribuido y paralelo. MapReduce divide las tareas de procesamiento en dos fases: *Map* y *Reduce*. La primera filtra y clasifica los datos, mientras que la segunda realiza operaciones de agregación y resumen. Este modelo permite manejar grandes volúmenes de datos en paralelo de manera eficiente, pues indica una lógica de procesamiento y escritura aplicaciones que transforman conjuntos de datos grandes en tareas que abarcan conjuntos más simples y manejables.

* **Hadoop Common:** Dicho componente incluye las bibliotecas y utilidades necesarias para los otros módulos de Hadoop. Proporciona servicios básicos y soporte para sistemas operativos, así como herramientas necesarias para la gestión y configuración del clúster de Hadoop.

En términos muy generales, la [arquitectura de Hadoop](https://hadoop.apache.org/docs/r1.2.1/hdfs_design.html) usa los componentes anteriores para 1) establecer una red entre múltiples computadoras y coordinar la comunicación entre ellas, 2) explotar la red anterior para construir un sistema distribuido tal que dado un conjunto de datos lo divida en un conjunto de bloques y los replique entre sus nodos, 3) aproveche lo anterior para ejecutar tareas de procesamiento de volumenes de datos en conjuntos más simples, 4) coordinar flujos de procesamiento de datos y asignar recursos para llevarlos acabo de forma exitosa.

## 2.3 ¿Qué es el ecosistema de Hadoop?

Además de los componentes básicos de Hadoop (HDFS, MapReduce, YARN y Hadoop Common), el [ecosistema de Hadoop](https://www.edureka.co/blog/hadoop-ecosystem) incluye varias herramientas y frameworks que facilitan el procesamiento y análisis de grandes volúmenes de datos. Entre los más destacados se encuentran:

* **Apache Pig:** es una plataforma de alto nivel para crear programas que se ejecutan en Hadoop. Su principal componente es *Pig Latin*, un lenguaje de scripting que permite a los usuarios escribir tareas de procesamiento de datos complejas de manera más sencilla que con MapReduce. Pig convierte estos scripts en series de trabajos de MapReduce, optimizando automáticamente su ejecución. Pig es especialmente útil para tareas de transformación, limpieza y carga de datos (ETL).

* **Apache Hive:** corrresponde un sistema de *data warehousing* construido sobre Hadoop, que facilita la consulta y análisis de grandes conjuntos de datos mediante un lenguaje similar a SQL llamado HiveQL. Hive traduce las consultas HiveQL en trabajos de MapReduce, haciendo que el procesamiento de datos en Hadoop sea más accesible para usuarios con conocimientos de SQL. Hive es adecuado para análisis de datos y generación de informes.

* **Apache HBase:** es una base de datos NoSQL distribuida y orientada a columnas que se ejecuta sobre HDFS. HBase permite almacenar grandes cantidades de datos estructurados y proporciona acceso aleatorio y en tiempo real a estos datos. Es ideal para aplicaciones que requieren operaciones de lectura y escritura rápidas y en gran volumen.

* **Apache Spark:** es un motor de procesamiento de datos de propósito general que es mucho más rápido que MapReduce, gracias a su capacidad para realizar procesamiento en memoria. Spark admite varios tipos de tareas de procesamiento de datos, incluidos batch, streaming, y análisis en tiempo real. Spark se integra con Hadoop y puede usar HDFS para el almacenamiento de datos.

* **Apache Flume:** se trata de un servicio para recopilar, agregar y mover grandes cantidades de datos de registros de múltiples fuentes a un almacenamiento centralizado como HDFS. Flume es muy utilizado para el análisis de datos de logs y el monitoreo en tiempo real.

* **Apache Sqoop:** es una herramienta diseñada para transferir datos entre sistemas de almacenamiento estructurados (como bases de datos relacionales) y Hadoop. Sqoop permite importar y exportar datos de manera eficiente, facilitando la integración de datos entre sistemas tradicionales y el ecosistema Hadoop.

* **Apache Oozie:** es un coordinador de flujos de trabajo y tareas en Hadoop. Oozie permite definir y ejecutar flujos de trabajo complejos que constan de múltiples trabajos de MapReduce, Pig, Hive, y otros. También soporta la programación de trabajos recurrentes y la gestión de dependencias entre tareas.

* **Apache ZooKeeper:** es un servicio centralizado para mantener información de configuración, nomenclatura, sincronización distribuida y servicios de grupo para grandes sistemas distribuidos. ZooKeeper es fundamental para la coordinación y gestión de clústeres de Hadoop.

Estos componentes, cuando se usan juntos, proporcionan un ecosistema robusto y flexible para el procesamiento y análisis de big data. Cada herramienta tiene un propósito específico y, en conjunto, facilitan la implementación de soluciones de big data que son escalables, eficientes y capaces de manejar diversos tipos de datos y cargas de trabajo.

## 2.4 ¿Cómo se puede ejecutar Hadoop?

Hay términos generales, hay 3 forma de [configurar](https://hadoop.apache.org/docs/stable/hadoop-project-dist/hadoop-common/SingleCluster.html#Standalone_Operation) dicha herramienta:

* **Modo Local:** El modo local es para cuando necesitas ejecutar un trabajo de Hadoop localmente en una sola máquina. Es útil para probar un trabajo de Hadoop y dar nuestros primeros pasos.

* **Modo Pseudo Distribuido:** Este modo simula un entorno de clúster distribuido en una sola máquina. Es útil para probar y validar un entorno distribuido en un solo dispositivo.

* **Modo Distribuido:** Relativo a la configuración de un clúster de máquinas distribuidas, desde unos pocos nodos hasta muchos más. Es el más adecuado para el entorno de producción. 

Sobre éste último punto cabe destacar que la mayoría de proveedores de cómputo en la nube ofrecen servicios que permiten la ejecución administrada de entornos de Hadoop, por ejemplo: 1) Amazon Web Services ofrece el servicios de [Elastic Map Reduce](https://aws.amazon.com/es/emr/), y 2) Google Cloud provee el servicio de [Dataproc](https://cloud.google.com/dataproc?hl=es-419)

## 2.5 Interactuando con los componentes de Hadoop

Una vez desplegado Hadoop, podemos interactuar con sus elementos como el HDFS, MapReduce y YARN a través de la linea de comandos del **Namenode** pues por diseño sigue una [arquitectura](https://hadoop.apache.org/docs/current/hadoop-project-dist/hadoop-hdfs/HdfsDesign.html) de tipo `master-slave`.


### 2.5.1 HDFS

En general, para interactuar a través del CLI con el sistema de archivos de Hadoop (HDFS se utiliza el comando `-dfs`. La lista de comandos se encuentra en la [documentación oficial](https://hadoop.apache.org/docs/current/hadoop-project-dist/hadoop-common/FileSystemShell.html), pero podemos mencionar que la sintaxis general es como sigue:

```
hdfs dfs -COMMAND [GENERIC_OPTIONS] [COMMAND_OPTIONS]
```

Ahora veremos ejemplos, pero siguiendo las [Notas de Aitor Medrano sobre Big Data](https://aitor-medrano.github.io/iabd/hadoop/hadoop.html) podemos mencionar algunos de los comandos más empleados en el día a día:

**Visualizando el contenido de un archivo del HDFS en la terminal**

**Otros**
* `count`: Cuenta el número de elementos (número de carpetas, ficheros, tamaño y ruta).

### 2.5.2 Hadoop JAR y Hadoop Streaming

### Hadoop Jar

Como sabemos, el framework Hadoop está implementado en Java por lo que es natural que se pueda emplear éste lenguaje de programación para realizar tareas personalizadas de procesamiento.

Java es un lenguaje de programación y una plataforma informática que fue desarrollado en la década de 1990. Es uno de los lenguajes de programación Orientado a Objetos y multiplataforma.

Hadoop posee funcionalidad `hadoop jar` que permirte correr en sus componentes al contenido de un archivo JAR (*Java ARchive*), que es una especie archivo comprimido que se utiliza en Java para agrupar múltiples archivos en uno solo, pues contiene archivos de clases (compilados en bytecode), metadatos y otros recursos necesarios para ejecutar una aplicación o librería Java.

Esta capacidad es muy interesante, pues permite hacer **ejecucciones personalizadas de procesamiento de datos** aprovechando al componente de **Map Reduce**.

La sintaxis de lo anterior es la siguiente:

```
hadoop jar my_program.jar [<arguments>]
````

#### Ejemplo: Ejecutando código personalizado en Hadoop Jar

Para ilustrar lo anterior, se presenta una version modificada de la [documentación de Hadoop](https://hadoop.apache.org/docs/stable/hadoop-mapreduce-client/hadoop-mapreduce-client-core/MapReduceTutorial.html) para un programa donde se hace el conteo de palabras de archivos de texto en un HDFS:

* Primero, se crea el código personalizado en un archivo `WordCount.java`:

```
// Codigo personalizado para hacer conteo de palabras en archivos de texto
// usando el framerwork de Map Reduce

// Importa librerias
import java.io.IOException;
import java.util.StringTokenizer;

import org.apache.hadoop.conf.Configuration;
import org.apache.hadoop.fs.Path;
import org.apache.hadoop.io.IntWritable;
import org.apache.hadoop.io.Text;
import org.apache.hadoop.mapreduce.Job;
import org.apache.hadoop.mapreduce.Mapper;
import org.apache.hadoop.mapreduce.Reducer;
import org.apache.hadoop.mapreduce.lib.input.FileInputFormat;
import org.apache.hadoop.mapreduce.lib.output.FileOutputFormat;

public class WordCount {

  // Mapper: clase que extiende la funcionalidad de Mapper para procesar la entrada
  public static class TokenizerMapper
       extends Mapper<Object, Text, Text, IntWritable>{

    // Definición de un valor constante para representar el conteo de uno
    private final static IntWritable one = new IntWritable(1);
    private Text word = new Text();

    // Método map: procesa una línea de texto y la divide en palabras
    public void map(Object key, Text value, Context context
                    ) throws IOException, InterruptedException {
      StringTokenizer itr = new StringTokenizer(value.toString());
      while (itr.hasMoreTokens()) {
        word.set(itr.nextToken());
        // Escribe cada palabra en el contexto con un conteo de uno
        context.write(word, one);
      }
    }
  }

  // Reducer: clase que extiende la funcionalidad de Reducer para agregar los valores
  public static class IntSumReducer
       extends Reducer<Text,IntWritable,Text,IntWritable> {
    private IntWritable result = new IntWritable();

    // Método reduce: suma los conteos de cada palabra
    public void reduce(Text key, Iterable<IntWritable> values,
                       Context context
                       ) throws IOException, InterruptedException {
      int sum = 0;
      for (IntWritable val : values) {
        sum += val.get();
      }
      result.set(sum);
      // Escribe el resultado en el contexto
      context.write(key, result);
    }
  }

  // Método main: configura y ejecuta el trabajo de MapReduce
  public static void main(String[] args) throws Exception {
    Configuration conf = new Configuration();
    Job job = Job.getInstance(conf, "word count");
    job.setJarByClass(WordCount.class);
    job.setMapperClass(TokenizerMapper.class);
    job.setCombinerClass(IntSumReducer.class);
    job.setReducerClass(IntSumReducer.class);
    job.setOutputKeyClass(Text.class);
    job.setOutputValueClass(IntWritable.class);
    
    // Configura las rutas de entrada y salida para los datos
    FileInputFormat.addInputPath(job, new Path(args[0]));
    FileOutputFormat.setOutputPath(job, new Path(args[1]));
    // Inicia el trabajo y espera a que termine
    System.exit(job.waitForCompletion(true) ? 0 : 1);
  }
}

```

* Posteriormente, se compila el programa anterior
 
```
# Instruccion para compilar el programa
hadoop com.sun.tools.javac.Main WordCount.java
```

Este programa genera una serie de archivos en el directorio entro los que se encuentra `wc.jar`

```
# Crea el archivo .jar
jar cf wc.jar WordCount*.class
```

* Ahora consideremos que tenemos la siguiente estructura en el hdfs:

```
hadoop fs -cat /tmp/wordcount/input/file01
Hello World Bye World

hadoop fs -cat /tmp/wordcount/input/file02
Hello Hadoop Goodbye Hadoop
```

* Ejecutando del jar obtendremos lo siguiente

```
# ejecutamos en archivo jar del codigo personalizado
hadoop jar wc.jar WordCount /tmp/wordcount/input /tmp/wordcount/output
```

Lo que induce el siguiente resultado:

```
hadoop fs -cat /tmp/wordcount/output
Bye 1
Goodbye 1
Hadoop 2
Hello 2
World 2
```

### Hadoop Streaming

Cómo hemos vista `hadoop jar` aprovecha las funcionalidades de Java. Este es un lenguaje que no necesariamente es dominado por todos en el ecosistema de Big Data, pero afortunadamente [Hadoop Streaming](https://hadoop.apache.org/docs/r3.3.1/hadoop-streaming/HadoopStreaming.html) permite crear y ejecutar tareas de MapReduce con cualquier script de MapReducer en el lenguahe de nuestra preferencia.

Esencialmente se trata de una utilidad donde creamos una tarea de map y reduce en un lenguajes o como resultado de un una aplicación y que toma el estandar input y output de la misma para procesar los datos (podemos pensar que es algo muy paracido a como funcionar los pipes de encadenamiento de instrucciones en UNIX: `comando_1 | comando_2 | comando_3 | ...`)

La sintaxis es como sigue:

```
mapred streaming \
  -input myInputDirs \
  -output myOutputDir \
  -mapper my_mapper_program_or_app \
  -reducer my_reducer_program_or_app
```

### Ejemplo: Hadoop Streaming con Python

El siguiente es un ejemplo modificado de la [notas de Big Data de Aitor Medrano](https://aitor-medrano.github.io/iabd/hadoop/hadoop.html#mapreduce-en-python) donde se construyen un mapper y reducer de Python para contar palabras de un texto:

* Archivo **mapper.py**

````
#!/usr/bin/python3

import sys

# En este loop se toma el texto del estandar input para
# procesarlo linea por linea

for linea in sys.stdin: 
    # Elimina los espacios de delante y detrás
    linea = linea.strip()

    # dividimos la línea en palabras
    palabras = linea.split()
    
    # creamos tuplas de (palabra, 1)
    for palabra in palabras: 
        print(palabra, "\t1")
```

* Archivo **reducer.py**

```
#!/usr/bin/python3 
import sys

# inicializamos el diccionario
dictPalabras = {}

# Este loop procesa las tuplas del mapper
# y suma los valores de las llaves para
# consolidar los conteos de palabras 

for linea in sys.stdin:
    # quitamos espacios de sobra
    linea = linea.strip()

    # parseamos la entrada de mapper.py, es decir,
    # dividimos en palabras separadas por saltos de linea
    palabra, cuenta = linea.split('\t', 1)

    # convertimos cuenta de string a int
    try:
        cuenta = int(cuenta)
    except ValueError:
        # Si no era un numero, descartamos la linea
        continue

    try:
        dictPalabras[palabra] += cuenta
    except:
        dictPalabras[palabra] = cuenta

for palabra in dictPalabras.keys():
    print(palabra, "\t", dictPalabras[palabra])
```

* Ahora podemos aplicar lo anterior:
 
```
mapred streaming \
    -input mi_texto_.txt \
    -output salida \
    -mapper mapper.py -file mapper.py \
    -reducer reducer.py -file reducer.py 
```

Lo cual redunda en lo siguiente:

```
hdfs dfs -head /salida/part-00000

hola      2
mundo     2
...
```
